## 0. 本文档作用

本文件用于作为 5 周课程的总控讲义，回答下面三个问题：

1. 这门课整体要讲什么。
2. 每一周在主线中承担什么角色。
3. 当前代码仓库中的哪些模块对应哪一周的内容。

## 1. 课程定位

### 1.1 项目定位

CityScholar-Agent 是一个面向城市治理、城市规划与城市安全方向的学术文献科研助教智能体。
课程以本地论文知识库为基础，围绕“最小智能体 -> 工具调用 -> LangChain/LangGraph -> RAG -> 评估”逐步展开。

### 1.2 课程目标

不是做一个单纯的聊天机器人，而是要理解并实现一个围绕科研任务运行的智能体原型。

课程结束后，学生至少应理解：

- 什么是最小智能体。
- 为什么工具调用是智能体区别于普通问答系统的重要能力。
- 为什么本地论文知识库是科研助教场景中的关键基础设施。
- 为什么评估必须和系统设计一起考虑，而不能最后再补。

### 1.3 课程主线

```text
-> 最小智能体
-> 工具调用
-> 流程编排（langchain、langgraph）
-> 本地知识库与 RAG
-> 评估与扩展
```

## 2. 总体智能体设计架构

### 2.1 设计原则

- 城市问题是主语，AI 与智能体是方法。
- 论文知识库是基础，科研任务完成是目标。
- 优先保证最小可运行，再逐步提升检索质量和任务复杂度。
- 所有模块都应可讲解、可演示、可继续迭代。

### 2.2 总体架构图

```mermaid
flowchart TD
    U[用户任务输入] --> R[任务识别与路由]
    R --> P[Planner 任务拆解]
    P --> S[State 状态管理]

    S --> T1[search_papers 检索论文]
    S --> T2[read_paper_chunks 读取片段]
    S --> T3[compare_papers 多篇比较]
    S --> T4[generate_outline 综述提纲]
    S --> T5[export_markdown 导出结果]

    T1 --> KB[(本地论文知识库)]
    T2 --> KB

    KB --> RET[Retriever 检索器]
    RET --> LLM[LLM 生成与总结]
    T3 --> LLM
    T4 --> LLM

    LLM --> O[结构化输出]
    O --> E[Evaluation 评估]
    E --> LOG[日志与复盘]
```

### 2.3 架构分层说明

#### 输入层

作用：接收用户问题、分析命令与研究任务。

当前对应文件：

- `app.py`

提示：

- 向学生强调：智能体的入口不一定复杂，但一定要清楚区分“接收输入”和“执行逻辑”。

#### 协调层

作用：负责把输入转化成具体任务，并协调底层模块完成工作。

当前对应文件：

- `core/agent.py`
- `core/prompts.py`

提示：

- 这是最适合讲“最小智能体”的位置。
- 当前版本虽然还不复杂，但已经体现了“调度者”的角色。

#### 知识处理层

作用：负责把原始论文转成可检索、可分析的数据对象。

当前对应文件：

- `rag/loader.py`
- `rag/parser.py`
- `rag/splitter.py`
- `rag/retriever.py`

提示：

- 这层是第 4 周 RAG 的自然铺垫。
- 第 1 周先让学生知道“知识处理链路存在”，不用一次讲透。

#### 工具层

作用：面向具体科研任务输出结构化结果。

当前对应文件：

- `tools/analyze_tool.py`

后续规划文件：

- `tools/compare_tool.py`
- `tools/outline_tool.py`
- `tools/export_tool.py`

提示：

- 第 2 周可以从这里自然过渡到“工具调用”的主题。

#### 输出层

作用：把结果整理为命令行展示、Markdown 文档或后续 notebook 结果。

当前对应文件：

- `app.py`
- `tools/analyze_tool.py`

## 3. 当前代码与课程主线的对应关系

### 3.1 已完成部分

- 本地 PDF 扫描与解析。
- 文本切块与最小知识库构建。
- 最小检索问答闭环。
- 单篇论文结构化提取原型。

### 3.2 明确属于后续任务

- 多篇论文对比。
- 综述提纲生成。
- embedding / 向量索引。
- 元数据过滤与混合检索。
- 评估脚本与测试问题集。

讲解提示：

- 可以明确告诉学生：当前版本已经足够支撑第 1 周教学。
- 但它还不是完整 V1，也不是课程最终版。
- 这种“先跑通、后增强”的路线本身就是课程要传递的工程思想。

## 4. 每周任务概况（简版）

## 第 1 周：最小科研助教智能体

### 本周目标

让学生理解什么是围绕科研任务工作的最小智能体，并跑通一个最小可运行版本。

### 本周关键词

- 最小闭环
- 智能体入口
- 任务导向
- 论文知识库原型

### 本周要讲的核心文件

- `app.py`
- `core/agent.py`
- `core/prompts.py`

### 本周课堂演示建议

```bash
python app.py
```

进入后依次演示：

```text
papers
analyze
哪些论文关注城市安全或韧性？
exit
```

### 本周产出

- 一个本地可运行的最小智能体原型。

---

## 第 2 周：工具调用与模块化

### 本周目标

让系统从“会答”走向“会做”，把结构化任务作为工具接入智能体。

### 本周关键词

- 工具调用
- 模块职责
- 结构化输出
- 多论文比较

### 本周要扩展的模块

- `tools/compare_tool.py`
- `tools/export_tool.py`
- `app.py` 中的工具命令分发

### 本周产出

- 一个可以调用多个科研工具的原型。

---

## 第 3 周：LangChain / LangGraph 与多步流程

### 本周目标

让学生理解当任务从单步变成多步时，为什么需要流程编排和状态管理。

### 本周关键词

- 多步任务
- 状态管理
- 流程编排
- 条件分支

### 本周建议讨论的问题

- 为什么“论文比较”比“单篇分析”更适合做多步流程示例。
- 为什么“综述提纲生成”天然适合链式或图式流程。

### 本周产出

- 一个带简单流程编排能力的增强版原型。

---

## 第 4 周：RAG 接入本地论文知识库

### 本周目标

把论文知识库真正并入智能体主线，而不是把 RAG 单独当成一个孤立话题。

### 本周关键词

- PDF 解析
- 切块
- embedding
- 向量索引
- Top-k 检索

### 本周建议新增模块

- `rag/embedder.py`
- `rag/knowledge_base.py`
- 可能的 `rag/reranker.py`

### 本周产出

- 一个更稳健的本地论文 RAG 助手。

---

## 第 5 周：评估、边界与扩展

### 本周目标

让学生学会从“能运行”走向“能解释、能评估、能改进”。

### 本周关键词

- 检索评估
- 回答评估
- 错误分析
- 系统边界
- 扩展路线

### 本周建议新增模块

- `eval/eval_retrieval.py`
- `eval/eval_answer.py`
- `eval/test_queries.json`

### 本周产出

- 一套可演示、可解释、可评估的课程原型系统。
